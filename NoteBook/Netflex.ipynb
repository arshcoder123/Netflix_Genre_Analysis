```python
import pandas as pd
```

```python
df = pd.read_csv("../CSV/NetFlix.csv")
```

```python
df= pd.DataFrame(df)
```

```python
df.info()
df.isna().sum()
df.shape
```

Data Cleaning

```python
df[["director","cast","country","rating"]]= df[["director","cast","country","rating"]].fillna("Unknown")
```

```python
df_og= df
```

```python
df["genres"] = df["genres"].str.split(",")
df["genres"] = df["genres"].apply(lambda x: [i.strip() for i in x])
```

```python
df_genre = df.explode("genres")
```

```python
df["cast"] = df["cast"].str.split(",")
df["cast"] = df["cast"].apply(lambda x: [i.strip() for i in x])
```

```python
df_cast = df.explode("cast")
```

```python
df["date_added"] = pd.to_datetime(df["date_added"])
```

Feature Engineering

```python
df["year_added"] = df["date_added"].dt.year
df["month_added"] = df["date_added"].dt.month
```

```python
df["year_added"] = df["year_added"].fillna(0).astype(int)
df["month_added"] = df["month_added"].fillna(0).astype(int)
```

Save File

```python
df.to_csv("netflix_cleaned.csv", index=False)
```
